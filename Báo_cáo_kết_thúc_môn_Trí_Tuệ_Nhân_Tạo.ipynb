Ở bảng dữ liệu này do có quá nhiều số liệu không thể kiểm soát được nên khó tránh khỏi việc có sai số, nên chúng em sẽ tiến hành lọc bảng dữ liệu này.


```python
#Dùng drop để xóa với axis=1 cột ('ID', 'Z_CostContact', 'Z_Revenue') và inplace=True để giữ DataFrame với các mục nhập hợp lệ trong cùng một biến.

df.drop(['ID', 'Z_CostContact', 'Z_Revenue'], axis=1, inplace=True)
```

```python
# Lọc giá trị thu nhập trung bình với fillna để lọc các giá trị phù hợp vào trong cột Income (thu nhập ).
df['Income'] = df['Income'].fillna(df['Income'].mean())
```

```python
# Kiểm tra xem ngày cuối cùng của dữ liệu này là bao nhiêu để dễ dàng lọc độ tuổi của khách hàng.
original_df['Dt_Customer'] = pd.to_datetime(original_df['Dt_Customer'])

print("The oldest record on customer's enrollment:", min(original_df['Dt_Customer']).date())
print("The newest record on customer's enrollment:", max(original_df['Dt_Customer']).date())
```

```python
## Lọc tuổi của khách hàng.
## Ở đoạn code trên có thể thấy ngày cuối cung là vào cuối năm 2014 nên chúng em sẽ lấy dữ liệu vào năm 2015 để làm chẳn.
df['Age'] = 2015 - df["Year_Birth"]
```

```python
# Thay thế một chuỗi các dữ liệu mối quan hệ bằng tình trạng hôn nhân thông qua replace.
df['Marital_Status'] = df['Marital_Status'].replace(['Married', 'Together'],'Relationship')
df['Marital_Status'] = df['Marital_Status'].replace(['Divorced', 'Widow', 'Alone', 'YOLO', 'Absurd'],'Single')
```

```python
# Thay thế các chuỗi dữ liệu về bằng cấp, chức vụ bằng trình độ học vấn.
df['Education'] = df['Education'].replace(['PhD','2n Cycle', 'Master'],'PG')
df['Education'] = df['Education'].replace(['Graduation', 'Basic'], 'UG')
```

```python
# Gộp dữ liệu của kidhome và teenhome vào childrenhome.
df['ChildrenHome'] = df['Kidhome'] + df['Teenhome']
# Gộp dữ liệu về các khoản chi trong dữ liệu(rựu,trái cây, thịt, cá đồ ngọt và mua vàng) vào khoản dữ liệu số tiền chi.
df['AmountSpent'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
# Gộp các dữ liệu chiến dịch vào cùng 1 dữ liệu TotalAcceptedCmp.
df['TotalAcceptedCmp'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] + df['Response']
# Gộp tổng số lần mua hàng từ các dữ liệu web,danh mục,cửa hàng, số giao dịch mua hàng.
df['NumTotalPurchases'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases'] + df['NumDealsPurchases']
```

```python
#Thời gian khách hàng duy trì mua hàng tại công ty đó.
df['Dt_Customer'] = pd.to_datetime(df.Dt_Customer)
df['first_day'] = '01-01-2015'
df['first_day'] = pd.to_datetime(df.first_day)
df['DaysEngaged'] = (df['first_day'] - df['Dt_Customer']).dt.days
```

```python
#Xóa đi các mục sau khi đã gộp dữ liệu.
cols_to_be_deleted = ["Kidhome", "Teenhome",
                      "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds",
                      "NumWebVisitsMonth", "NumWebPurchases","NumCatalogPurchases","NumStorePurchases","NumDealsPurchases",
                      "AcceptedCmp1" , "AcceptedCmp2", "AcceptedCmp3" , "AcceptedCmp4","AcceptedCmp5", "Response",
                      "Dt_Customer", "first_day", "Year_Birth", "Dt_Customer", "Recency", "Complain"]

df=df.drop(columns=cols_to_be_deleted, axis=1)
```

```python
## sắp xếp lại các cột sau khi đã xóa các danh mục được lọc ở trên
df = df[['Age', 'Education', 'Marital_Status', 'ChildrenHome', 'Income', 'TotalAcceptedCmp', 'AmountSpent', 'NumTotalPurchases', 'DaysEngaged']]
df.head()
```

```python
#sắp xếp tiêu đề của các hàng và cột dưới dạng danh sách theo thứ tự hàng chính
fig=make_subplots(rows=2,cols=2,subplot_titles=('<i>Age', '<i>Income', '<i>Amount Spent', '<i>Days Engaged'))
fig.add_trace(go.Box(x=df['Age'],name='Age'),row=1,col=1) #định dạng lưới ô của tuổi
fig.add_trace(go.Box(x=df['Income'],name='Income'),row=1,col=2) #thu nhập
fig.add_trace(go.Box(x=df['AmountSpent'],name='AmountSpent'),row=2,col=1) #số tiền chi
fig.add_trace(go.Box(x=df['DaysEngaged'],name='DaysEngaged'),row=2,col=2) #ngày đính hôn

fig.update_yaxes(visible=False, showticklabels=False)
fig.update_layout(height=600, width=1000, title_text='<b>Xử lý các giá trị ngoại lệ', font_size=20)
fig.update_layout(template='plotly_dark', title_x=0.5, font_family='Courier New', showlegend=False)
```

```python
## Lọc và loại bỏ những giá trị khác với phần còn lại của tuổi và thu nhập
df = df[df['Age'] < 100]
df = df[df['Income'] < 150000]
```

### 1.  Tình trạng hôn nhân của khách hàng?

Ở dữ liệu đầu tiên chúng em sẽ xem xét đến vấn đề là hững khách hàng đã kết hôn hoặc độc thân có chên lệch nhiều trong việc mua hàng tại cửa hàng này không và số liệu đó có đáng kể không để có được kết luận.

```python
sns.histplot(data=df,x= "Marital_Status")
## Nhóm lại các dữ liệu để tạo danh mục marital_count
marital_count = df.groupby(['Marital_Status']).size().reset_index().rename(columns={0: 'count'})
fig = make_subplots(rows=2, cols=2,
                    specs=[[{'type':'domain'}, {'type':'domain'}],
                           [{'type':'domain'}, {'type':'domain'}]
                          ])

## sắp xếp biểu đồ
fig.add_trace(
    go.Pie(
        labels=marital_count['Marital_Status'],
        values=marital_count['count'],
        hole=.6,
        title='Marital Status',
        titlefont={'color':None, 'size': 10},
        ),

#Cập nhật vào biểu đồ
    )
fig.layout.update(title="<b> Biểu đồ về tình trạng hôn nhân <b>",
                  titlefont={'color':None, 'size': 28, 'family': 'Courier New'},
                  showlegend=False,
                  height=800,
                  width=1000,
                  template='plotly_dark',
                  title_x=0.5
                  )
fig.show()
```

Dựa vào kết biểu đồ, ta có thể thấy khách hàng đã kiết hôn mua hàng nhiều hơn lượng khách hàng chưa kết hôn. Phần lớn lượng khách hàng đã kết hôn có nhu cầu chi tiêu cao hơn so với mảng còn lại.

### 4. Mối tương quan giữa số tuổi với thu nhập, số tiền tiêu dùng,số lần mua hàng và ngày đăng ký tại cửa hàng

> Liệu rằng ở từng độ tuổi khác nhau thì nhu cầu của khách hàng sẽ có thay đổi so với nhu cầu mua hàng và khả năng thu nhập của họ. Sơ đồ hồi quy này sẽ biểu diễn cụ thể hơn về nghi vấn đó.



```python
#tạo dòng biểu đồ, với dạng hồi quy tuyến tính x phù hợp với mảng để dự đoán y
def makeLine(df, x, y):
    model = LinearRegression().fit(np.array(df[x]).reshape(-1,1), (np.array(df[y])))
    y_pred = model.predict(np.array(df[x]).reshape(-1,1))
    return y_pred
```

```python
fig = make_subplots(
    rows=2, cols=2, subplot_titles=("Age vs Income", "Age vs Amount Spent", "Age vs Number of Purchases", "Age vs Days Engaged")
) #tạo ô phụ với 2 dòng 2 cột tương ứng với mỗi tiêu đề được niêu

# Add traces từ hàm "from  plotly.subplots  import  make_subplots"
# fig.add_trace() để cập nhật số liệu từ các nhóm thuộc tính của khách hàng
# Scatter mỗi điểm dữ liệu được biểu diễn dưới dạng một điểm đánh dấu, có vị trí được cung cấp bởi các cột xvà y
fig.add_trace(go.Scatter(x=df['Age'], y=df['Income'], mode='markers'), row=1, col=1)
fig.add_trace(go.Scatter(x=df['Age'], y=makeLine(df, 'Age', 'Income'), mode='lines',name="Linear_reg_fit", marker_color='white'), row=1, col=1)

fig.add_trace(go.Scatter(x=df['Age'], y=df['AmountSpent'], mode='markers'), row=1, col=2)
fig.add_trace(go.Scatter(x=df['Age'], y=makeLine(df, 'Age', 'AmountSpent'), mode='lines',name="Linear_reg_fit", marker_color='white'), row=1, col=2)


fig.add_trace(go.Scatter(x=df['Age'], y=df['NumTotalPurchases'], mode='markers'), row=2, col=1)
fig.add_trace(go.Scatter(x=df['Age'], y=makeLine(df, 'Age', 'NumTotalPurchases'), mode='lines',name="Linear_reg_fit", marker_color='white'), row=2, col=1)

fig.add_trace(go.Scatter(x=df['Age'], y=df['DaysEngaged'], mode='markers'), row=2, col=2)
fig.add_trace(go.Scatter(x=df['Age'], y=makeLine(df, 'Age', 'DaysEngaged'), mode='lines',name="Linear_reg_fit", marker_color='white'), row=2, col=2)


# Cập nhật xaxis thuộc tính "age" và sử dụng title.text để gép 2 trục x và y
fig.update_xaxes(title_text="Age", row=1, col=1)
fig.update_xaxes(title_text="Age", row=1, col=2)
fig.update_xaxes(title_text="Age", row=2, col=1)
fig.update_xaxes(title_text="Age", type="log", row=2, col=2)

# Cập nhật tiêu đề và chiều cao
fig.update_layout(title_text="Các thuộc tính với Age", title_x=0.5, height=800, template='plotly_dark', showlegend=False,
        font=dict(
            family="Rubik",
            size=14)
)

fig.show()
```
